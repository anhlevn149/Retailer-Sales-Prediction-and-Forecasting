# 2. Data Preprocessing

```python
df_clean = merged_df_price.copy()
```

```python
# merged_df_clean['event_name'] = merged_df_clean['event_name'].replace(np.nan, 'Not_Applicable', regex=True)
```

```python
# merged_df_clean['event_type'] = merged_df_clean['event_type'].replace(np.nan, 'Not_Applicable', regex=True)
```

```python
df_clean['sell_price'] = df_clean['sell_price'].replace(np.nan, 0, regex=True)
```

```python
df_clean.isna().sum()
```

```python
# Create a column named sales_revenue
df_clean['sales_revenue'] = df_clean['sales_qty'] * df_clean['sell_price']
```

```python
df_clean = df_clean.drop(['d', 'wm_yr_wk', 'sales_qty', 'sell_price'], axis=1)
df_clean.head()
```

```python
# Convert the 'date' column to a datetime type
df_clean['date'] = pd.to_datetime(df_clean['date'])
df_clean.head()
```

```python
import datetime as dt
```

```python
df_clean['year'] = df_clean['date'].dt.year
df_clean['quarter'] = df_clean['date'].dt.quarter
df_clean['month'] = df_clean['date'].dt.month
df_clean['day_of_week'] = df_clean['date'].dt.dayofweek  # 0=Monday, 1=Tuesday, ..., 6=Sunday

df_clean
```

```python
# Set date column to be dataframe index
df_clean.set_index('date')
```

```python
# Convert the 'date' column to a datetime type
pd.to_datetime(df_clean.index)
```

```python
# Import LabelEncoder from sklearn.preprocessing
from sklearn.preprocessing import LabelEncoder
```

```python
# Instantiate the LabelEncoder and save as le
le = LabelEncoder()
```

```python
# Create two new columns to fit and transform the label encoded columns ('item_id', 'store_id')
df_clean['item_id_le'] = le.fit_transform(df_clean['item_id'])
df_clean['store_id_le'] = le.fit_transform(df_clean['store_id'])
```

```python
df_clean = df_clean.drop(['item_id', 'store_id'], axis=1)
```

# 3. Train/ Validation Split

```python
# Split df_clean to training and validation sets with approximately 80:20 ratio
train = df_clean.loc[df_clean['date'] < '14-06-2014']
val = df_clean.loc[df_clean['date'] >= '14-06-2014']
```

```python
# Create a list called features, including following columns: 'item_id', 'store_id', 'year', 'quarter', 'month', 'day_of_week'
features = ['item_id_le', 'store_id_le', 'year', 'quarter', 'month', 'day_of_week']

# Create a list called target, including following columns: 'sales_revenue'
target = ['sales_revenue']
```

```python
# Save as X_train, y_train, X_val, y_val
X_train = train[features]
y_train = train[target]

X_val = val[features]
y_val = val[target]
```

# 4. Data scale

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Instantiate the StandardScaler method and save as a variable called scaler
scaler = StandardScaler()
```

```python
# Fit the training dataset to the scaler
scaler.fit(X_train)
```

```python
# Apply the scaler to the training, and validation datasets to replace current values of features
X_train = scaler.transform(X_train)
X_val = scaler.transform(X_val)
```
